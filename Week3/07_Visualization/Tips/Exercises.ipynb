This exercise was created based on the tutorial and documentation from [Seaborn](https://stanford.edu/~mwaskom/software/seaborn/index.html)  
The dataset being used is tips from Seaborn.

### Step 1. Import the necessary libraries:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv"
data = pd.read_csv(url)
```

### Step 3. Assign it to a variable called tips

```python
tips = data

print(tips.head())
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop(columns=["Unnamed: 0"], inplace=True)
print(tips.head())
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], bins=20, kde=True)  
plt.title('Distribution of Total Bill Amounts in Tips Dataset', fontsize=14, pad=15)  
plt.xlabel('Total Bill ($)', fontsize=12)  
plt.ylabel('Count', fontsize=12)  
plt.grid(True, linestyle='--', alpha=0.7)  
plt.xticks(fontsize=10)  
plt.yticks(fontsize=10)  
plt.tight_layout()  
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', color='coral')
plt.title('Total Bill vs Tip Relationship', fontsize=14, pad=15)
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', size='size', sizes=(20, 200), color='teal')
plt.title('Total Bill vs Tip by Size', fontsize=14, pad=15)
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='day', palette='Set2', legend=False)
plt.title('Total Bill Distribution by Day', fontsize=14, pad=15)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='tip', y='day', hue='sex', palette='Set1')
plt.title('Tip Amount by Day Differentiated by Sex', fontsize=14, pad=15)
plt.xlabel('Tip ($)', fontsize=12)
plt.ylabel('Day', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='time', palette='coolwarm')
plt.title('Total Bill per Day Differentiated by Time (Dinner or Lunch)', fontsize=14, pad=15)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)
plt.legend(title='Time', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.figure(figsize=(14, 6))

# Histogram for Dinner
plt.subplot(1, 2, 1)
sns.histplot(tips[tips['time'] == 'Dinner']['tip'], bins=20, kde=True, color='blue')
plt.title('Tip Distribution for Dinner', fontsize=14, pad=15)
plt.xlabel('Tip ($)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Histogram for Lunch
plt.subplot(1, 2, 2)
sns.histplot(tips[tips['time'] == 'Lunch']['tip'], bins=20, kde=True, color='green')
plt.title('Tip Distribution for Lunch', fontsize=14, pad=15)
plt.xlabel('Tip ($)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
plt.figure(figsize=(14, 6))

# Scatterplot for Male
plt.subplot(1, 2, 1)
sns.scatterplot(data=tips[tips['sex'] == 'Male'], x='total_bill', y='tip', hue='smoker', palette='coolwarm')
plt.title('Total Bill vs Tip (Male)', fontsize=14, pad=15)
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.legend(title='Smoker', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Scatterplot for Female
plt.subplot(1, 2, 2)
sns.scatterplot(data=tips[tips['sex'] == 'Female'], x='total_bill', y='tip', hue='smoker', palette='coolwarm')
plt.title('Total Bill vs Tip (Female)', fontsize=14, pad=15)
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.legend(title='Smoker', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### BONUS: Create your own question and answer it using a graph.

```python
# Question: What is the average tip amount for each day of the week, and how does it compare?

plt.figure(figsize=(10, 6))
avg_tips_per_day = tips.groupby('day')['tip'].mean().reset_index()
sns.barplot(data=avg_tips_per_day, x='day', y='tip', hue='day', palette='viridis', legend=False)
plt.title('Average Tip Amount by Day', fontsize=14, pad=15)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Average Tip ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
